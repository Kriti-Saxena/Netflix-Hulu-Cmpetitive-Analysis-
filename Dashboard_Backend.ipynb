```python
hulu_genre=list(data_hulu['listed_in'])

string=[]


for x in hulu_genre:
    spl=x.split(', ')
    for y in spl:
        string.append(y)

string=pd.DataFrame(string)
string.columns=['genre']
group= string.groupby('genre').agg({'genre':'count'})
group.apply(lambda x: x.sort_values(ascending=False))
```

```python
hulu_movies_genre = list(hulu_movie['listed_in'])

string_movies=[]


for x in hulu_movies_genre:
    split_movies=x.split(', ')
    for y in split_movies:
        string_movies.append(y)

string_movies=pd.DataFrame(string_movies)
string_movies.columns=['genre']
group= string_movies.groupby('genre').agg({'genre':'count'})
group.apply(lambda x: x.sort_values(ascending=False)) 
```
